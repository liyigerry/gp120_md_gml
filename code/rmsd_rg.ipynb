```python
def drawRMSD1(datapath,tied_name_list,tier,color,save_path,character):
    figure(figsize=(10,6))
    colors = color
    for tied1a_name in tied_name_list:
        top_h = md.load(datapath+tier+'/'+tied1a_name+'/'+tied1a_name+'_5fyk.pdb')
        traj = md.load('D:/Lab/data/gp120/5roles/data/trajs_5fyj/'+tier+'/'+tied1a_name+'/'+'pro.xtc',  top=top_h)
        rg=md.compute_rg(traj)
        plot(rg, color=next(colors),linewidth=1, label=tied1a_name)
    font={'size':12,
         'family':'Times New Roman',
         'weight':'bold',}
    legend(ncol=5,loc='upper center',prop=font)
    plt.text(2900, 2.51, '%s'%(character), family='times new roman',weight='bold')
    xlim(0,3000)
    xticks(range(0,3100,500),range(0,31,5))
    xlabel('Time (ns)', family='times new roman',weight='bold')
    ylim(2.5,3.0)
    yticks([2.5,2.6,2.7,2.8,2.9,3.0])
    ylabel('Rg (nm)', family='times new roman',weight='bold')
    savefig(save_path,format='png',dpi=600)
```
